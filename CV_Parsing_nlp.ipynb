### What is CV Parsing in NLP ? 
CV parsing uses natural language processing (NLP) techniques to extract structured information from unstructured CVs or resumes. It involves analyzing the text of a CV to identify and categorize key details such as personal information, work experience, education, skills, and other relevant data. This process helps in automating the recruitment process by allowing employers to quickly and efficiently screen and sort through large volumes of CVs.

### CV Parsing Basics 
Resume parsing automates information extraction from resumes, making it easier for recruiters to manage and analyze candidate data. The process typically involves the following steps:
1. **Document Analysis**: The parser reads the CV document, which can be in various formats such as PDF, Word, or plain text.
2. **Text Extraction**: The parser extracts the raw text from the document, removing any formatting or layout elements.
3. **Information Extraction**: The parser identifies and categorizes key information such as:
   - Personal details (name, contact information)
   - Work experience (job titles, company names, dates)
   - Education (degrees, institutions, graduation dates)
   - Skills (technical skills, soft skills)
4. **Data Structuring**: The extracted information is organized into a structured format, such as a database or a standardized template, making it easier for recruiters to search and filter candidates based on specific criteria.

### Why Use CV Parsing ?
CV parsing offers several benefits for both recruiters and job seekers:
- **Time Efficiency**: Automates the process of reviewing resumes, allowing recruiters to focus on qualified candidates.
- **Improved Accuracy**: Reduces human error in data entry and ensures consistent formatting of candidate information.
- **Enhanced Candidate Experience**: Provides a more streamlined application process for job seekers, allowing them to easily submit their resumes without worrying about formatting issues.
- **System Integration**: Can be integrated with applicant tracking systems (ATS) to further streamline the recruitment process and improve candidate management.

### The Role of NLP in CV Parsing
NLP plays a crucial role in CV parsing by enabling the extraction of meaningful information from unstructured text. NLP techniques such as named entity recognition (NER), part-of-speech tagging, and dependency parsing help in identifying and categorizing different elements of a CV. For example, NER can be used to identify names, job titles, and company names, while part-of-speech tagging can help in understanding the context of certain phrases and sentences. 
 - Recognizing synonyms and variations in language is also important in CV parsing, as candidates may use different terms to describe the same skills or experiences. NLP can help in normalizing this information to ensure that it is accurately categorized and searchable.
 - Understanding the context of certain phrases and sentences is also crucial in CV parsing, as it can help in determining the relevance of certain information to a particular job role. For example, a candidate may mention a skill or experience that is not directly related to the job they are applying for, but may still be relevant in terms of transferable skills or potential for growth. NLP can help in identifying these nuances and providing a more comprehensive understanding of a candidate's qualifications.
 - Identifying and categorizing different elements of a CV is also important in CV parsing, as it allows recruiters to quickly and easily search for specific information. For example, a recruiter may want to search for candidates with a certain level of education or specific technical skills. NLP can help in categorizing this information in a way that makes it easily searchable and filterable.

### How NLP-Based CV Parsing Works
- **Document Preprocessing**: The CV is preprocessed to clean and normalize the text. This may involve removing special characters, converting text to lowercase, and tokenizing the text into individual words or phrases.
- **Section Recognition**: The parser identifies different sections of the CV, such as personal information, work experience, education, and skills. This can be done using rule-based approaches or machine learning models trained on labeled CV data.
- **Entity Extraction**: NLP techniques are used to extract specific entities from the text, such as names, job titles, company names, dates, and skills. This may involve using named entity recognition (NER) models or custom-built extraction rules.
- **Data Normalization**: The extracted information is normalized to ensure consistency. For example, different variations of a skill (e.g., "Python programming" vs. "Python") may be standardized to a single term.
- **Structured Output**: The extracted and normalized information is organized into a structured format, such as a JSON object or a database entry, which can be easily accessed and analyzed by recruiters.

### NER with SpaCy: Core concept

 NER identifies and classifies named entities in text into predefined categories such as names, organizations, locations, dates, and more.
- **Data Annotation**:\

- **Custom labels**:\

### Practical Implementation: A Step-by-Step Guide

```python
import spacy
import re

### Load spacy model
nlp = spacy.load("en_core_web_sm")

## Sample resume text
resume_text = """
John Doe
Email: johndoe@example.com
Phone: +1 (123) 456-7890

Experienced in software development with a focus on Python and Java. 
Skilled in machine learning, data analysis, and web development.
"""
## Predefined skills list
SKILLS = ["Python", "Java", "machine learning", "data analysis", "web development"]

## Preprocess the text  with spacy
doc = nlp(resume_text)

### Extracting entities using NER
name = next((ent.text for ent in doc.ents if ent.label_ == "PERSON"), None)

### Exracting email using regex
email_match = re.search(r'[\w\.-]+@[\w\.-]+', resume_text)
email = email_match.group(0) if email_match else None

## Extract phone number using regex
phone_match = re.search(r'(\+?\d{1,3})?[\s\-]?\(?\d{3}\)?[\s\-]\d{3}[\s\-]?\d{4}', resume_text)
phone = phone_match.group(0) if phone_match else None

### Extracting skills
skills = [skill for skill in SKILLS if skill.lower() in resume_text.lower()]

## Print the extracted information
print(f"Name: {name}")
print(f"Email: {email}")
print(f"Phone: {phone}")
print(f"Skills: {', '.join(skills)}")



```
