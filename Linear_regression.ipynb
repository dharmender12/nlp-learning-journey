### ML crash course: Linear Regression 


#### Linear Regression 
Linear regression is a fundamental machine learning algorithm used for predicting a continuous target variable based on one or more input features. The goal of linear regression is to find the best-fitting line (or hyperplane in higher dimensions) that describes the relationship between the input features and the target variable. The linear regression model can be expressed mathematically as:
y = β0 + β1x1 + β2x2 + ... + βnxn + ε

Where:
- y is the target variable (the value we want to predict).
- x1, x2, ..., xn are the input features (the independent variables).
- β0 is the intercept (the value of y when all x's are 0).
- β1, β2, ..., βn are the coefficients (the weights assigned to each input feature).
- ε is the error term (the difference between the predicted and actual values). 

### Fundamental of Linear Regression
#### What is Regression?
Regression is a type of supervised machine learning algorithm used for predicting a continuous target variable based on one or more input features. The goal of regression is to find the relationship between the input features and the target variable, allowing us to make predictions on new, unseen data. Regression models can be linear or non-linear, depending on the nature of the relationship between the input features and the target variable. Linear regression is a specific type of regression that assumes a linear relationship between the input features and the target variable.

#### Simple Linear Regression
Simple linear regression is a type of linear regression that involves only one input feature (independent variable) and one target variable (dependent variable). The model can be expressed as:

y = β0 + β1x + ε

Where:
- y is the target variable (the value we want to predict).
- x is the input feature (the independent variable).
- β0 is the intercept (the value of y when x is 0).
- β1 is the coefficient (the weight assigned to the input feature).
- ε is the error term (the difference between the predicted and actual values).

#### Ordinary Least Squares (OLS)
Ordinary Least Squares (OLS) is a method used to estimate the coefficients of a linear regression model. The OLS method minimizes the sum of the squared differences between the predicted values and the actual values. The OLS estimates for the coefficients can be calculated using the following formulas:

β1 = (Σ((x_i - x̄)(y_i - ȳ))) / (Σ((x_i - x̄)^2))

β0 = ȳ - β1x̄

Where:
- x_i and y_i are the individual data points for the input feature and target variable, 
- x̄ and ȳ are the mean values of the input feature and target variable, respectively.   


### Model Evaluation Metrics
#### R-squared (R²) 
R-squared (R²) is a statistical measure that represents the proportion of the variance in the target variable that is explained by the input features in the linear regression model. R² values range from 0 to 1, where a value of 0 indicates that the model does not explain any of the variance in the target variable, and a value of 1 indicates that the model explains all of the variance. A higher R² value indicates a better fit of the model to the data.

#### Adjuted R-squared
Adjusted R-squared is a modified version of R-squared that takes into account the number of input features in the model. It is used to prevent overfitting by penalizing the addition of irrelevant features. Adjusted R-squared can be calculated using the following formula: 
Adjusted R² = 1 - [(1 - R²)(n - 1) / (n - k - 1)]

Where:
- R² is the R-squared value of the model.
- n is the number of observations in the dataset.
- k is the number of input features in the model.

#### Error Metrics
- Mean Absolute Error (MAE): MAE is the average of the absolute differences between the predicted values and the actual values. It is calculated as:
MAE = (1/n) * Σ|y_i - ŷ_i|
Where:- y_i is the actual value of the target variable for the i-th observation.
- ŷ_i is the predicted value of the target variable for the i-th observation.
- n is the total number of observations in the dataset.


### Practical Moddel Building
#### Data Preprocessing
Data preprocessing is a crucial step in building a linear regression model. It involves cleaning and transforming the data to ensure that it is in a suitable format for modeling. Common data preprocessing steps include handling missing values, encoding categorical variables, and feature scaling. Handling missing values can be done by either removing the rows with missing values or imputing them using techniques such as mean, median, or mode imputation. Encoding categorical variables can be achieved using techniques like one-hot encoding or label encoding. Feature scaling is important to ensure that all input features are on the same scale, which can be done using methods like standardization or normalization.

#### Train-Test Split
Train-test split is a technique used to evaluate the performance of a linear regression model. It involves splitting the dataset into two subsets: a training set and a testing set. The training set is used to fit the linear regression model, while the testing set is used to evaluate the model's performance on unseen data. A common split ratio is 80% for training and 20% for testing, but this can be adjusted based on the size of the dataset and the specific requirements of the problem.

#### Model Fitting 
Model fitting is the process of estimating the coefficients of the linear regression model using the training data. This can be done using various optimization algorithms, such as gradient descent or ordinary least squares (OLS). The goal of model fitting is to find the values of the coefficients that minimize the error between the predicted values and the actual values in the training data.

#### Validation & Refinement 
Validation is the process of evaluating the performance of the linear regression model on the testing set. This involves calculating evaluation metrics such as R-squared, adjusted R-squared, mean absolute error (MAE), and mean squared error (MSE) to assess how well the model generalizes to unseen data. Based on the validation results, the model can be refined by adjusting the input features, adding interaction terms, or applying regularization techniques to improve its performance and prevent overfitting.

```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score    
import matplotlib.pyplot as plt
import seaborn as sns

### 2. Load the Dataset

dataset = pd.read_csv('salary_data.csv')

print("---- First 5 rows of the dataset ----")
print(dataset.head())
print("\n")

## 3. Data Preprocessing
x = dataset[['YearsExperience']]
y = dataset['Salary']

### 4. Train-Test Split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)   

### 5. Initialize and Train the Model
model = LinearRegression()
### 6. Model Fitting
model.fit(x_train, y_train)

print('--- Model Training Completed ---')
print("\n")

### 7. Understanding the Model Coefficients
print(f"Intercept (β0): {model.intercept_}")
print(f"Coefficient (β1): {model.coef_[0]}")
print("The equation of the Model is: Salary = {:.2f} + {:.2f} * YearsExperience".format(model.intercept_, model.coef_[0]))
print("\n")

### 8. Making Predictions
y_pred = model.predict(x_test)

### 9. Model Evaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared Score: {r2:.2f}")
print("The R-squared score indicates that {:.2f}% of the variance in Salary can be explained by Years of Experience.".format(r2 * 100))
print("R-square tells us how they are related to each other. The closer the R-squared value is to 1, the better the model fits the data.")

print("\n")
print("The Mean Squared Error (MSE) is a measure of the average squared difference between the predicted and actual values. A lower MSE indicates that the model's predictions are closer to the actual values, which is desirable for a good model fit.")
print("\n")
```
