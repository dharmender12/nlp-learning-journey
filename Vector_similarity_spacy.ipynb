### Vector Similarity in Spacy 
vector similarity is a fundamental concept in natural language processing (NLP) that measures the similarity between two vectors, typically representing the semantic meaning of words or documents.



### Introduction to Word Embeddings
- What are Word Vectors?

Numerical representation of words in a multi-dimensional space where semantic similar words have similar vector representation.

- Semantic Relationship:

Words with similar meaning have cluster together in the vector space, capturing semantic and syntactic relationships.

### Vector Similarity Core concepts 
1. Similarity Score
    - A float between 0-1, higher means more similar
2. Cosine Similarity
    - Measures the angle between two vectors
    - Range: -1.0 to 1.0
    - 1.0: Identical vectors
    - 0.0: Orthogonal vectors
    - -1.0: Opposite vectors
3. .similarity() method :
    - Available on Doc, Span, and Token objects
    - Returns a similarity score on a scale of -1.0 to 1.0
4. Model Selection
    - Must use medium or larger models with vectors
    - spaCy uses pre-trained models for vectorization
    - Models are selected based on the language and domain
    - Common models: en_core_web_sm, de_core_news_sm

### Token Similarity

**Comparing Individual Words**

Token similarity is a measure of how similar two words are based on their vector representation. It is a way to compare the semantic meaning of two words.

The example below shows how to compare the similarity between two words using spaCy's token similarity.


### Document and Span Similarity 

1. **Document Similarity**
Document similarity measures how similar two documents are based on the average of the word vectors in each document
2. **Span Similarity**
Span similarity measures the similarity between two spans of text, which can be a phrase, sentence,or any contiguous sequence of tokens. It is calculated by averaging the vectors of the tokens in each span and then computing the cosine similarity between the two average vectors.

3. **Mixed Comparisons**
You can also compare a token to a document or a span to a document. The similarity will be calculated based on the average vector of the document or span and the vector of the token.

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
sentence1 = 'I love playing football.'
# sentence2 = 'I enjoy playing soccer.'
sentence2 = 'I am going to the mall.'
doc1 = nlp(sentence1)
doc2 = nlp(sentence2)

similarity = doc1.similarity(doc2)
print(f"Sentence 1: {sentence1}")
print(f"Sentence 2: {sentence2}")
print(f"Similarity Score: {similarity:.4f}")
```

## Similarity between two words

```python
word1 = nlp('cat')
word2 = nlp('dog')
similarity = word1.similarity(word2)
print(f"Word 1: {word1.text}")
print(f"Word 2: {word2.text}")
print(f"Similarity Score: {similarity:.4f}")
```

### Limitations of Vector Similarity
1. Context and Negation
    - Vector similarity may not capture the context or negation in sentences, leading to inaccurate similarity scores.
2. Out-of-Vocabulary Words
    - Words that are not present in the model's vocabulary will not have vector representations, resulting in zero similarity scores.
3. Domain-Specific Language
    - Pre-trained models may not perform well on domain-specific language, leading to lower similarity scores for relevant terms.   
4. Averaging effects
    - Averaging word vectors for documents or spans can dilute the meaning and lead to less accurate similarity scores, especially for longer texts.
