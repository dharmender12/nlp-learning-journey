### Overview: AI Poetry Generation
1. **Algorithmic Approach**:
    - Combines natural language processing (NLP) techniques with deep learning models to generate poetry. These models learn from large datasets of existing poems to understand the structure, style, and themes commonly found in poetry.
2. **Style Mimicry**:
    - AI models can mimic the style of specific poets or genres by training on their works. This allows for the creation of poems that resemble the writing style of famous poets like Shakespeare, Emily Dickinson, or contemporary poets.
3. **Multiple Methods**: 
    - Approaches to poetry generation include rule-based systems, Markov chains, and more advanced deep learning models like Recurrent Neural Networks (RNNs) and Transformers. Each method has its own strengths and limitations in terms of creativity and coherence.
