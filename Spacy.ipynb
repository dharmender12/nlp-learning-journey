### POS tagging using Spacy
POS tagging is the process of assigning a part of speech tag to each word in a sentence. Spacy provides a lot of data about the tokens. We can access it using the token object.

- **token.text:** gives the original text of the token
- **token.pos_:** gives the coarse-grained part of speech tag
- **token.tag_:** gives the fine-grained part of speech tag
- **spacy.explain(token.tag_)**: gives the explanation of the fine-grained part of speech tag.

```python
# text = "the quick brown fox jumps over the lazy dog"
text = "I am Saara and I work at ABC. I am 80 year old. I like apples. I love reading books."
doc = nlp(text)
print(f"{'token':<15}{'POS':<5}{'detailed tag':<18}{'Explanation'}") ## header for the table, < means left align and 15 means width of the column
# print(f"{'Token'}{'POS'}{'Detailed Tag'}{'Explanation'}") ## header for the table, without alignment
print("*"*60) ## separator line, we can use any character like - or = or * etc. and we can adjust the number of characters to match the width of the table
for token in doc: 
    print(f"{token.text:<15}{token.pos_:<10}{token.tag_:<15}{spacy.explain(token.tag_)}")
```

### Dependency parsing using Spacy
Dependency parsing is the process of analyzing the grammatical structure of a sentence and establishing relationships between "head" words and words which modify those heads. Spacy provides a lot of data about the tokens. We can access it using the token object.
- **token.text:** gives the original text of the token
- **token.dep_:** gives the dependency label of the token
- **token.head.text:** gives the text of the head token
- **token.head.pos_:** gives the coarse-grained part of speech tag of the head token
- **token.head.tag_:** gives the fine-grained part of speech tag of the head token
- **spacy.explain(token.dep_)**: gives the explanation of the dependency label.

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
text = "The quick brown fox jumps over the lazy dog."
doc = nlp(text)
print(f"{'token':<15}{'Head':<12}{'Dependency':<15}{'Head POS':<10}{'Explanation':<20}{'Explanation dep'}") ## header for the table, < means left align and 15 means width of the column
print("*"*100) ## separator line, we can use any character like - or = or * etc. and we can adjust the number of characters to match the width of the table
for token in doc:
    print(f"{token.text:<15}{token.head.text:<12}{token.dep_:<15}{token.head.pos_:<10}{spacy.explain(token.tag_):<35}{spacy.explain(token.dep_)}")
```

### Lemmatization using Spacy
---
Lemmatization is the process of reducing a word to its base form. Spacy provides a lot of data about the tokens. We can access it using the token object.
- **token.text:** gives the original text of the token
- **token.lemma_:** gives the lemma of the token
- **token.pos_:** gives the coarse-grained part of speech tag of the token
- **token.tag_:** gives the fine-grained part of speech tag of the token
- **spacy.explain(token.tag_)**: gives the explanation of the fine-grained part of speech tag.

### Difference between lemmatization and stemming
**Lemmatization** is the process of reducing a word to its base form, while 
**stemming** is the process of reducing a word to its root form. The main difference between lemmatization and stemming is that lemmatization considers the context and the meaning of the word, while stemming does not. Lemmatization is more accurate than stemming, but it is also more computationally expensive. Stemming is faster than lemmatization, but it can produce incorrect results.



```python
text = "The children are running faster than yesterday."
doc = nlp(text)
print(f"{'token':<15}{'Lemma':<12}{'POS':<10}{'Explanation'}") ## header for the table, < means left align and 15 means width of the column
print("*"*100) ## separator line, we can use any character like - or = or * etc. and we can adjust the number of characters to match the width of the table
for token in doc:
    print(f"{token.text:<15}{token.lemma_:<12}{token.pos_:<10}{spacy.explain(token.pos_)}")
```

### NER (Named Entity Recognition) using Spacy
NER is a task of identifying and classifying named entities in text into pre-defined categories. 

Spacy is a popular library for NLP tasks. It provides pre-trained models for NER which can be used for various NER tasks.


```python
nlp = spacy.load("en_core_web_sm")
```

```python
text = "Apple was founded by Steve Jobs and is headquartered in California."
doc = nlp(text)

print(f"{'Entity Text':<15} {'Entity Label':<15}{'Explanation'}")
print("-"*50)
for entity in doc.ents:
    print(f"{entity.text:<15} {entity.label:<15} {spacy.explain(entity.label_)}")
```
