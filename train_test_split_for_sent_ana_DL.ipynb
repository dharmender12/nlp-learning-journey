### What is Sentiment Analysis?
Sentiment Analysis, also known as Opinion Mining, is a Natural Language Processing (NLP) technique. It classifies text into categories such as positive, negative, or neutral based on the sentiment expressed. It is widely applied to understand opinions in : 
- social media monitoring
- customer feedback analysis
- Product reviews analysis

### Why split Data ? The Overfitting Problem: 
- **The Overfitting challenge**:\
  Models can memorize training examples rather than learn general patterns, leading to poor performance on unseen data. This is known as overfitting.
   - Model performs well on training data
   - same model performs poorly on new, unseen data or example data 
   - Without proper splits, evaluation results become misleading, giving a false sense of model performance.

### Strandard Data Splitting Ratios:
- **Training Set**:\
    - Used to fit the model parameters through the learning process. 
    - Typically the largest portion of the dataset, often around 70-80% of the total data.
- **Validation Set**:\
    - Used to tune hyperparameters and make decisions about model architecture.
    - Helps in preventing overfitting by providing an unbiased evaluation during training.
    - Commonly around 10-15% of the total data.
- **Test Set**:\
    - Used for the final evaluation of the model's performance after training is complete.
    - Provides an unbiased assessment of the model's generalization to new data.
    - Typically around 10-15% of the total data.


### Typical Split Ratios in Deep Learning:
- **Standard Split**:
    - 70% Training, 15% Validation, 15% Test
- **Validation Data**:
    - 80% Training, 10% Validation, 10% Test
- **Test Data**:
    - 60% Training, 20% Validation, 20% Test

### How to Perform a Split: 
- **Use Framework Tools**:
    - Libraries like Scikit-learn provide functions such as `train_test_split` to easily split datasets into training, validation, and test sets.
- **Random Shuffling**:
    - Ensure that the data is randomly shuffled before splitting to avoid any bias in the distribution of classes across the sets.
- **Stratification**:
    - For NLP tasks, especially with imbalanced datasets, use stratified sampling to maintain the same class distribution in each split.

### Example : IMDB Sentiment Analysis Dataset:
- 50K - Total Reviews
- 25K - Training Set
- 25K - Test Set
- No separate validation set provided, so we can create one from the training set (e.g., 20K for training and 5K for validation).

### Best Practices and Mistakes to Avoid:
- **Keep Test Set Separate**:\
  Always keep the test set completely separate from the training and validation sets to ensure an unbiased evaluation of model performance.
- **Prevent Data Leakage**:\
  Avoid any form of data leakage where information from the test set is inadvertently used during training or validation, as this can lead to overly optimistic performance estimates.
- **Maintain Class Balance**:\
  Ensure that the class distribution is maintained across all splits, especially in cases of imbalanced datasets, to prevent biased model performance.
