### Logistic Regression
#### What is Logistic Regression?
- Classification Algorithm : 
 A supervised learning algorithm used for binary classification problems.
- Predicts Probability : Predicts the probability of an instance belonging to a particular class (usually binary, such as 0 or 1).
- Uses Logistic Function : Utilizes the logistic function (sigmoid function) to map predicted values to a range between 0 and 1, which can be interpreted as probabilities.
- Mathematical Expression : The logistic regression model can be expressed mathematically as:
p = 1 / (1 + e^-(β0 + β1x1 + β2x2 + ... + βnxn))    

Where:
- p is the predicted probability of the instance belonging to the positive class (class 1).
- x1, x2, ..., xn are the input features (the independent variables).
- β0 is the intercept (the log-odds of the positive class when all x's are 0).
- β1, β2, ..., βn are the coefficients (the weights assigned to each input feature).

- Desicion Boundary : The decision boundary for logistic regression is typically set at a threshold (e.g., 0.5), where instances with predicted probabilities above the threshold are classified as the positive class, and those below are classified as the negative class.

 

### Linear vs Logistic Regression
| Aspect                  | Linear Regression                        | Logistic Regression                      |
|-------------------------|-----------------------------------------|-----------------------------------------|
| Type of Problem         | Regression (predicting continuous values) | Classification (predicting class    labels) |
| Output                  | Continuous values (e.g., price, temperature) | Probabilities (between 0 and 1) that can be mapped to class labels |
| Model Equation          | y = β0 + β1x1 + β2x2 + ... + βnxn + ε | p = 1 / (1 + e^-(β0 + β1x1 + β2x2 + ... + βnxn)) |
| Loss Function           | Mean Squared Error (MSE)                     | Log Loss (Cross-Entropy Loss)           |
| Decision Boundary         | Not applicable (predicts continuous values) | Typically set at a threshold (e.g., 0.5) to classify instances into classes |

### Evaluation Metrics for Logistic Regression
- **Accuracy :** \
The proportion of correctly classified instances out of the total instances.
- Precision : \
 The proportion of true positive predictions out of all positive predictions (TP / (TP + FP)).
- Recall (Sensitivity) : \
 The proportion of true positive predictions out of all actual positive instances (TP / (TP + FN)).
- F1 Score: \
 The harmonic mean of precision and recall, providing a single metric that balances both (2 * (Precision * Recall) / (Precision + Recall)).
- ROC Curve & AUC : \
 The area under the Receiver Operating Characteristic curve, which measures the model's ability to distinguish between classes across different threshold values. A higher AUC indicates better performance.
- Confusion Matrix : \ 
 A table that summarizes the performance of a classification model by showing the counts of true positives, true negatives, false positives, and false negatives. It helps to visualize the performance of the model and identify areas for improvement.

### Extensions of Logistic Regression
- **Binary Logistic Regression :** \
 The standard form of logistic regression used for binary classification problems. Example, either class 0 or class 1.
- **Multinomial Logistic Regression :** \
    An extension of logistic regression used for multi-class classification problems, where the target variable can take on more than two classes. It models the probabilities of each class using a softmax function. \
    Multiple unordered classes (e.g., class 0, class 1, class 2).
- **Ordinal Logistic Regression :** \
Multiple ordered classes (e.g., low, medium, high). \
An extension of logistic regression used for ordinal classification problems, where the target variable has a natural order but the distances between classes are not necessarily equal. It models the cumulative probabilities of the classes using a logistic function.

```python
### 1. Import necessary libraries

import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
## 2. Load the dataset
dataset = pd.read_csv('Social_Network_Ads.csv')
print("---First 5 rows of the dataset---")
print(dataset.head())
print("\n")

### 3. prepare the data 
### x contains the features (independent variables) 
x = dataset[['Age','EstimatedSalary']]
### y contains the target variable (dependent variable)
y = dataset.iloc[:, -1].values

### 4. split the data into training and testing sets
## Random state is set to 0 to ensure reproducibility of results
## Random state is the seed used by the random number generator, 
# which ensures that the same sequence of random numbers is generated each time the code is run. 
# This allows for consistent results when splitting the data into training and testing sets.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
### 5. Feature Scaling
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

### 6. Create and train the Logistic Regression model
## Initialize the Logistic Regression model
model = LogisticRegression(random_state=0)
## Fit the model to the training data
model.fit(x_train, y_train)

print("---Model training completed---")
print("\n")

### 7. Make predictions using the trained model
y_pred = model.predict(x_test)

### 8. Evaluate the model's performance
print("---Model Evaluation---")

## Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
print("\n")

## Classification Report
cr = classification_report(y_test, y_pred)
print("Classification Report:")
print(cr)
print("\n")

### Interpretation of results
## The confusion matrix shows the number of true positives, true negatives, false positives, and false negatives. 
## The classification report provides precision, recall, f1-score, and support for each class. 
## Precision is the ratio of true positives to the sum of true positives and false positives
```
