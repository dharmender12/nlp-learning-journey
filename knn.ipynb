## K-Nearest Neighbors (KNN) : 
#### A simple yet effective Machine Learning Algorithm
K-Nearest Neighbors (KNN) is a supervised machine learning algorithm that can be used for both classification and regression tasks. It is a non-parametric method, meaning it does not make any assumptions about the underlying data distribution. KNN works by finding the K nearest neighbors to a given data point and making predictions based on the majority class (for classification) or the average value (for regression) of those neighbors.

### Core Concept: You are Like Your Neighbors
The core idea behind KNN is that similar data points are likely to be close to each other in the feature space. The algorithm assumes that data points that are close to each other are more likely to belong to the same class (for classification) or have similar values (for regression). The KNN algorithm uses a distance metric (such as Euclidean distance) to measure the similarity between data points and identify the nearest neighbors.

1. **Tell me who your friends are, and I will tell you who you are**: \
Imagine you are trying to classify a new data point. You look around and find the K nearest neighbors to that data point. If most of those neighbors belong to a certain class, you can infer that the new data point likely belongs to that class as well. For example, if you are trying to classify a new email as spam or not spam, you would look at the K nearest emails in your dataset. If most of those emails are labeled as spam, you would classify the new email as spam too.

2. **Majority Vote Classification**: \
In classification tasks, KNN makes predictions based on the majority class among the K nearest neighbors. For instance, if K=3 and two of the nearest neighbors belong to class A while one belongs to class B, the algorithm would classify the new data point as class A.

3. **Average Value Regression**: \
In regression tasks, KNN predicts the value of a new data point by taking the average of the values of the K nearest neighbors. For example, if you are trying to predict the price of a house based on its features, you would look at the K nearest houses in your dataset and take the average price of those houses to make your prediction.



### How KNN Works: The Algorithm Steps
1. **Choose the number of neighbors (K)**: \The first step in KNN is to choose the number of neighbors (K) that you want to consider when making predictions. The choice of K can significantly impact the performance of the algorithm. A small K value may lead to a noisy decision boundary, while a large K value may smooth out the decision boundary too much.

2. **Calculate the distance**: \
Next, you need to calculate the distance between the new data point and all the data points in your training dataset. Common distance metrics include Euclidean distance, Manhattan distance, and Minkowski distance.

3. **Identify the nearest neighbors**: \ 
After calculating the distances, you need to identify the K nearest neighbors to the new data point. This involves sorting the distances and selecting the K data points with the smallest distances.
4. **Make a prediction**: \
Finally, you make a prediction based on the majority class (for classification) or the average value (for regression) of the K nearest neighbors. This is done by looking at the labels of the K nearest neighbors and determining which class is most common (for classification) or calculating the average value (for regression).

### Ways to calculate number of neighbors (K)
1. **Distance Calculation**: \
This is also known as **Euclidean Distance**. It is the most common method for calculating the distance between data points in KNN. The distance between two data points \( x \) and \( y \) can be calculated using various distance metrics. For example, the Euclidean distance is given by: \
$d(x, y) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}$ 

where \( n \) is the number of features.

2. **Manhattan Distance**: 
This distance metric calculates the distance between two points by summing the absolute differences of their coordinates. It is given by:\
$d(x, y) = \sum_{i=1}^{n} |x_i - y_i|$ 

where, \( n \) is the number of features. Manhattan distance is often used when the data points are located on a grid, such as in image processing or when dealing with categorical data.

3. **Minkowski Distance**: \
This is a generalization of both Euclidean and Manhattan distances. It is defined as: \
$d(x, y) = \left( \sum_{i=1}^{n} |x_i - y_i|^p \right)^{1/p}$ 

where \( p \) is a parameter that determines the type of distance. When \( p = 2 \), it is equivalent to Euclidean distance, and when \( p = 1 \), it is equivalent to Manhattan distance.



### Advantages and Limitations of KNN
**Advantages**:
- Simple to understand and implement.
- No training phase, making it computationally efficient for small datasets.
- Can be used for both classification and regression tasks.
- Non-parametric, meaning it does not make any assumptions about the underlying data distribution.
- Versatile and can be used for various types of data, including numerical and categorical data.

**Limitations**:
- Computationally expensive for large datasets, as it requires calculating the distance to all training data points.
- Sensitive to irrelevant features and the scale of the data, which can affect the performance of the algorithm.
- The choice of K can significantly impact the performance, and there is no definitive method for selecting the optimal K value.
- Can struggle with imbalanced datasets, where one class is significantly more prevalent than the other, leading to biased predictions.
- Does not perform well with high-dimensional data due to the curse of dimensionality, where the distance between data points becomes less meaningful as the number of features increases.

### Practical Implementation Steps
1. **Select Optimal K Value**: \
Use cross-validation to determine the optimal K value for your dataset. This involves splitting the data into training and validation sets and evaluating the performance of the KNN algorithm for different K values.
2. **Calculate Distances**: \
Choose an appropriate distance metric (e.g., Euclidean, Manhattan) based on the nature of your data and calculate the distances between the new data point and all training data points.
3. **Find Nearest Neighbors**: \
Identify the K nearest neighbors based on the calculated distances and their corresponding labels.
4. **Make Predictions**: \
For classification tasks, use majority voting to determine the predicted class. For regression tasks, calculate the average value of the K nearest neighbors to make the prediction.
5. **Evaluate Performance**: \
Evaluate the performance of the KNN algorithm using metrics such as accuracy, precision, recall, F1-score for classification, and mean squared error for regression. Use cross-validation to ensure that the model generalizes well to unseen data.

### Real-World Applications of KNN
1. **Image Recognition**: \KNN can be used for image classification tasks, where the algorithm identifies the class of an image based on the classes of its nearest neighbors in the feature space.
2. **Recommender Systems**: \KNN can be used in recommender systems to suggest products or content based on the preferences of similar users (user-based KNN) or similar items (item-based KNN).
3. **Medical Diagnosis**: \KNN can be applied in medical diagnosis to classify diseases based on patient data, where the algorithm predicts the disease based on the symptoms and medical history of similar patients.
4. **Customer Segmentation**: \KNN can be used for customer segmentation in marketing, where customers are grouped based on their purchasing behavior and preferences, allowing businesses to target specific segments with personalized marketing strategies.
5. **Anomaly Detection**: \KNN can be used for anomaly detection in various domains, such as fraud detection in finance or network security, where the algorithm identifies outliers based on the distance to their nearest neighbors in the feature space.

```python
### 1. Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

### 2. Define the data manually 

data = {
    'customerID': [1, 2, 3, 4, 5],
    'age': [25, 30, 35, 40, 45],
    'income': [50000, 60000, 70000, 80000, 90000],
    'spending_score': [60, 70, 80, 90, 100]
}

### 3. Create a DataFrame from the data
df = pd.DataFrame(data)

### 4. Select the features for clustering
X = df[['age', 'income', 'spending_score']]

### 5. Apply K-Means Clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)

### 6. Print the DataFrame with cluster labels
df['cluster'] = kmeans.labels_

### 7. Print the DataFrame with cluster labels
print(df)

### 8. Visualize the clusters (2D plot: Age vs Spending Score)
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['spending_score'], c=df['cluster'], cmap='viridis',s=100)
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.title('Customer Segmentation using K-Means Clustering')
plt.colorbar(label='Cluster')
plt.grid()
plt.show()





```
