### Introduction to Hate Speech Classification:
Hate speech classification is a critical task in natural language processing (NLP) that involves identifying and categorizing harmful language in text data. This task is essential for maintaining safe online environments and combating the spread of hate speech on social media platforms, forums, and other digital spaces.\
The modern internet faces a growing challenge: identifying and addressing hate speech.\
As online platforms expand, so does the need for effective tools to detect and mitigate harmful content.\
This presentation explores how we classify harmful content to create safer digital spaces.

### What is Hate Speech?
Hate speech comprises language that expresses prejudice, against protected groups.\
It creates a division through "us vs. them" framing.
- **Target Characteristics**:\
    - Race vs. ethinicity.
    - Religion vs. belief
    - Gender identify
    - Sexual orientation
    - Disability

### Categories and Intensity of Hate Speech:
1. **Early Warning**:
    - Stereotyping and oversimplification of groups, which can lead to the formation of negative perceptions and biases.
2. **Dehumanization**:
    - Language that portrays individuals or groups as less than human, often using derogatory terms and imagery to justify discrimination and violence.
3. **Incitement**:
    - Speech that encourages or incites violence or discrimination against individuals or groups, often through direct calls to action or by creating a hostile environment.
    - Call for violence, discrimination, or exclusion of targeted groups.

### Examples of Hate Speec in Practice:
1. **Historical Context**:
    - The Rwandan Genocide in 1994 was fueled by hate speech broadcasted through radio, which dehumanized the Tutsi minority and incited violence.
2. **Social Media**:
    - Stereotyping posts that reinforce harmful group perceptions, such as those targeting specific ethnic or religious groups, can contribute to a toxic online environment and perpetuate discrimination.
3. **Calls to Action**:
    - Messages urging protests, exclusion, or violence against specific groups can lead to real-world harm and escalate tensions within communities.

### Why Classify Hate Speech?
- **Platform Safety**:\
    Classifying hate speech helps online platforms maintain a safe and inclusive environment for users by identifying and removing harmful content.
- **User Protection**:\
    It protects individuals and groups from harassment, discrimination, and violence by enabling timely intervention against hate speech.
- **Early Intervention**:\
    Early detection of hate speech can prevent escalation and mitigate the spread of harmful content, fostering healthier online communities.

### Datasets and Data Sources:
1. **Platform Sources**:
    - Social media platforms (e.g., Twitter, Facebook) provide a wealth of user-generated content that can be used for hate speech classification.
2. **Labelling Methods**:
    - Manual annotation by human experts to create labeled datasets for training and evaluating hate speech classifiers.
3. **Balanced Sampling**:
    - Diverse content across demographics, topics, and severity levels to ensure the model learns to recognize various forms of hate speech effectively.

### Features Used in Classification:
1. **Textual Features**:
    - words, phrases, sentiment analysis, contextual cues, and linguistic patterns that indicate hate speech.
2. **Metadata Features**:
    - User profile information, posting time, location data and engagement metrics that can provide additional context for classification.
3. **Protected Characteristics**:
    - References to specific groups or identify markers that can help the model recognize targeted hate speech.


### Machine Learning Approaches:
1. **Traditional Machine Learning**:
    - Algorithms like Support Vector Machines (SVM), Naive Bayes, and Logistic Regression that rely on handcrafted features for classification.
2. **NLP Techniques**:
    - Natural Language Processing techniques such as tokenization, stemming, and lemmatization to preprocess text data for better feature extraction.
    - TF-IDF and word embeddings (e.g., Word2Vec, GloVe) to represent text data in a format suitable for capture meaning.
3. **Deep Learning Models**:
    - Neural networks, including Convolutional Neural Networks (CNNs) and Recurrent Neural Networks (RNNs), that can automatically learn complex patterns in text data for improved classification performance.
    - Transformer models like BERT understand context and nuances in language, making them particularly effective for hate speech classification.

### Conclusion and Future Directions:
- **Ongoing Challenges**:
    - Addressing the evolving nature of hate speech, including new slang and coded language, to ensure classifiers remain effective.
- **Transparency Requirements**:
    - Ensuring that hate speech classification models are transparent and explainable to build trust with users and stakeholders.
- **Ethical Implementation**:
    - Balancing the need for effective hate speech detection with the protection of free speech and avoiding over-censorship.

The future of hate speech classification lies at the intersection of AI capabilities and human values.

**Text Cleaning Steps**:
1. **Raw Text**:
    - Original text data as collected from sources, which may contain noise and irrelevant information.
2. **Basic Cleaning**:
    - Removing URLs, mentions, hashtags, and special characters to reduce noise in the data.
3. **Tokenization**:
    - Breaking down text into individual words or tokens for analysis.
4. **Stemming and Lemmatization**:
    - Reducing words to their base or root form to standardize the text for better feature extraction.
    - Reducing words to their root forms using NLTK tools.

### Baseline Machine Learning Models:
1. **Logistic Regression**:
    - A simple linear model that can be effective for binary classification tasks like hate speech detection.
    - Fast training time 
    - Interpretable coefficients that indicate feature importance.
    - Works well with TF-IDF features and can serve as a strong baseline for comparison with more complex models.
2. **Naive Bayes**:
    - A probabilistic model based on Bayes' theorem, often used for text classification tasks due to its simplicity and efficiency.
    - Handles high-dimensional data well, making it suitable for text classification with large vocabularies.
    - Works well with small to medium-sized datasets and can be effective for hate speech detection when combined with appropriate feature extraction techniques.
    - Fast prediction time, making it suitable for real-time applications.
3. **Decision Trees**:
    - Rule-based models that can capture non-linear relationships in data, but may require careful tuning to avoid overfitting.
    - Typical accuracy ~ 70-80% on hate speech datasets, depending on the quality of features and the dataset size.
    - Highly interpretable results, allowing for easy understanding of the decision-making process.
    - Can capture non-linear patterns in the data, but may require pruning to prevent overfitting.


Models are trained on vectorised text features and corresponding hate speech labels, allowing them to learn patterns and make predictions on new, unseen data.

### Text Collection and Initial Cleaning:
1. **Data Collection**:
    - Gathering text data from various sources, such as social media platforms, forums, and news comments, to create a comprehensive dataset for hate speech classification.
2. **Text Normalization**:
    - Converting text to lowercase, removing punctuation, and eliminating stop words to standardize the data for better feature extraction.
3. **Noise Removal**:
    - Eliminating irrelevant information, such as URLs, mentions, and special characters, to reduce noise in the data and improve model performance.

### Tokenisation & Normalization: 
1. **Tokenization**:
    - Breaking down text into individual words or tokens for analysis.
2. **Stopword Removal**:
    - Eliminating common words that do not contribute to the meaning of the text, such as "the", "is", and "and", to focus on more informative features for classification.
3. **Hash Tag Preservation**:
    - Retaining hashtags as they can provide valuable context and information for hate speech classification, especially on social media platforms where they are commonly used to convey sentiment and group affiliations.

### Lemmatisation and stemming:
1. **Word Reduction**:
    - Reducing words to their base or root form to standardize the text for better feature extraction and improve model performance.
2. **Tool Implementation**:
    - Utilizing tools like NLTK's WordNetLemmatizer for lemmatization and PorterStemmer for stemming to preprocess text data effectively for hate speech classification.
3. **Dimensionality Benefits**:
    - Reducing the number of unique words in the dataset, which can help improve model performance by focusing on the core meaning of words rather than their various forms.

### Handling Code - Mixed and Multilingual Text:
1. **Language Identification**:
    - Implementing language detection algorithms to identify the language of the text, which is crucial for effective preprocessing and classification in multilingual datasets.
2. **Standardization**:
    - Developing strategies to standardize and normalize text across different languages and scripts to ensure consistent feature extraction and model performance in multilingual hate speech classification.
3. **Custom Tokenisers**:
    - Creating custom tokenization approaches to handle code-mixed text, which is common in multilingual contexts, to improve the accuracy of hate speech classification models.

### Feature Extraction and Vectorization:
1. **Basic Vectorization**:
    - Converting text data into numerical representations using techniques like Bag of Words and TF-IDF to enable machine learning models to process and learn from the data.
2. **Advanced Embeddings**:
    - Utilizing word embeddings (e.g., Word2Vec, GloVe) and contextual embeddings (e.g., BERT) to capture semantic meaning and context in text data for improved classification performance.
3. **Data Splitting**:
    - Dividing the dataset into training, validation, and test sets to evaluate model performance and prevent overfitting.
