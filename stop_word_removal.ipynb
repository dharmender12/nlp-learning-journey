### Lemmatisation and Stemming
Lemmatisation and stemming are techniques used in natural language processing (NLP) to reduce words to their base or root form.
- **Stemming**: This technique involves cutting off the end of a word to reduce it to its base form. For example, "running" would be reduced to "run". However, stemming can sometimes produce non-dictionary words (e.g., "happily" might be reduced to "happil").
- **Lemmatisation**: This technique is more sophisticated than stemming. It reduces words to their base or dictionary form (lemma) while considering the context. For example, "running" would be reduced to "run", and "better" would be reduced to "good". Lemmatisation typically requires a dictionary and is more accurate than stemming, but it can be computationally more expensive.

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
words = ['running', 'ran', 'runs', 'easily', 'fairly']
stemmed_words = [ps.stem(word) for word in words]
print("Original Words:", words)
print("Stemmed Words:", stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
words = ['running', 'ran', 'runs', 'easily', 'fairly']
lemmatized_words = [lemmatizer.lemmatize(word) for word in words] # pos='v' for verb, default is noun
print("Original Words:", words)
print("Lemmatized Words:", lemmatized_words)
```

### Name Entity Recognition (NER)
Name Entity Recognition (NER) is a subtask of information extraction that seeks to locate and classify named entities mentioned in unstructured text into predefined categories such as the names of persons, organizations, locations, expressions of times, quantities, monetary values, percentages, etc.

```python
nltk.download('averaged_perceptron_tagger') # For POS tagging
nltk.download('maxent_ne_chunker') # For NER
nltk.download('words') # For NER
```

```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_eng')
nltk.download('maxent_ne_chunker_tab')
text = "Obama was born on August 4 in Honolulu."

tokenized_text = word_tokenize(text)
pos_tagged_text = pos_tag(tokenized_text)
named_entities = ne_chunk(pos_tagged_text)
print(named_entities)
```

## Part of Speech (POS) Tagging
Part of Speech (POS) tagging is the process of marking up a word in a text as corresponding to a particular part of speech, based on both its definition and its context. Common POS tags include nouns, verbs, adjectives, adverbs, pronouns, conjunctions, prepositions, and interjections. POS tagging is essential for various NLP tasks, including parsing, information extraction, and machine translation.

### Text Cleaning & Preprocessing using NLTK

```python
import nltk 
from nltk import FreqDist
from nltk.tokenize import word_tokenize

nltk.download('punkt_tab')
text = "This is a sample sentence. The sentence has some common words and some rare words."

tokens = word_tokenize(text.lower())


freq_dist = FreqDist(tokens)
filtered_words = [word for word in tokens if freq_dist[word] > 1]

print(filtered_words)
```
